# Load the tomoscopes

```python
folderPath = '/eos/project/l/liu/PS/data/2017/MD2586_InjectionEmittanceBlowup/2018.10.24_new_WS+BGI_brightness_curve/'
files_tomo = sorted(glob.glob(folderPath+'tomo_data/*.dat'))
```

```python
tomo_data = pnd.DataFrame()

for f in range(len(files_tomo)):
    try:
        tomo_d = myToolbox.extractProfile(files_tomo[f])
        tomo_d['Filename'] = files_tomo[f]
        if tomo_d is np.nan:
            pass
        else:
            tomo_data=tomo_data.append(tomo_d,ignore_index=True)
    except IOError:
        print('\nIOError\n')
    except IndexError:
        print('\nIndexError\n')
```

```python
plt.plot(1e-10*tomo_data['eperimage'],1e3*tomo_data['deltaP_P_RMS'],'.')
plt.xlabel('Intensity (1e10 ppb)')
plt.ylabel('dp/p_rms (1e-3)')
```

```python
# maybe a subset with no correlation ?
xx = 1e-10*tomo_data['eperimage'].loc[tomo_data['eperimage']>60e10]
yy = 1e3*tomo_data['deltaP_P_RMS'].loc[tomo_data['eperimage']>60e10]

fig, ax = plt.subplots(1,2, figsize=(14,7))
ax[0].plot(xx, yy, '.')
ax[0].set_xlabel('Intensity (1e10 ppb)')
ax[0].set_ylabel('dp/p_rms (1e-3)')

slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(xx, yy)
print( 'r^2 = '+str(r_value**2 ))

#let's have a look to the residuals
ax[1].plot(xx, yy-(slope*xx + intercept),'.')
ax[1].set_ylabel('Fit residuals')
ax[1].set_xlabel('Intensity (1e10 ppb)')

fig.tight_layout()
```

# The moment of truth

```python
# kill a failed tomo shot
tomo_data = tomo_data.drop(tomo_data.loc[120].name)
```

```python
fig, ax = plt.subplots(1,2,figsize=(14,7))
xx = tomo_data['BCT_intensity']
yy = 1e-10*tomo_data['eperimage']

ax[0].plot(xx, yy, '.')
ax[0].set_xlabel('BCT Intensity (1e10 ppb)')
ax[0].set_ylabel('TOMO Intensity (1e10 ppb)')

# fitting 
def lin(x, a, b):
    return a + b*x

popt, pcov = curve_fit(lin, xx, yy ); slope=popt[1]; intercept=popt[0]
sig_a = np.sqrt(np.diag(pcov))[0]
sig_b = np.sqrt(np.diag(pcov))[1]
sig_ab = pcov[0][1]

ax[0].plot(xx, xx*slope+intercept)
ax[0].text(60, 40, 'Linear fit: $a + bx$:\n $a=$%.3f +/- %.3f \n $b=$%.3f +/- %.3f'%(intercept,sig_a,slope,sig_b))

ax[1].plot(xx, yy-(xx*slope+intercept),'.')
ax[1].set_xlabel('BCT Intensity (1e10 ppb)')
ax[1].set_ylabel('Fit residuals (1e10 ppb)')


fig.tight_layout()
```
